6. A startup founder tells you — "I want to build a food delivery app like Zomato. I don't want to buy expensive servers and computers. Is there any other way?"

-  You don’t need to buy expensive servers or computers to build a food delivery app. You can use cloud computing.

      Cloud computing means using servers, storage, and software over the internet instead of owning them physically.Companies like Amazon Web Services, Microsoft Azure, and Google Cloud provide these services.

      You don’t need to set up servers or data centers Pay only for what you use Easy to scale High reliability Data is stored safely with backup

      Let’s say you build an app like Zomato :

      User opens your app and orders food

      Request goes to your app (hosted on cloud server)

      Cloud server processes the order

      Data (user, restaurant, order) is stored in cloud database

      Response is sent back to user instantly
      All this happens without you owning any physical server

      Conclusion :
      Cloud computing allows startups to build and run apps without heavy investment. It is cost-effective, scalable, and perfect for growing businesses.

7.  What is the difference between a compiler and an interpreter? Explain in your own words with a simple real-life comparison.
  
    - Compiler :
     A compiler translates the entire program a once into machine code
     After compiling, you can run the program
     Errors are shown after the whole code is checked
  
      Interpreter :
      An interpreter translates the program line by line It executes each line immediately
      Errors are shown as soon as they occur

      Conclusion :
      A compiler processes the entire code at once, while an interpreter works step by step. Both help convert human-readable code into machine language, but they work in different ways.


8. What is the Software Development Life Cycle (SDLC)? Name all its phases and write one line about what happens in each phase.
- Software Development Life Cycle (SDLC)
 SDLC is a step-by-step process used to develop high-quality software in a systematic way.

   Phases of SDLC (with one-line explanation)
  Planning Decide the project goals, scope, budget, and timeline Requirement Analysis Collect and understand what the user/client actually needs
  Design :
     Create the structure of the system (architecture, UI, database design)

  Development (Coding) :
     Developers write the actual code for the software

  Testing :
     Check for bugs and ensure the software works correctly
  Deployment :
     Release the software to users (make it live)

  Maintenance :
     Fix bugs, update features, and improve performance after release

  10. What is the difference between HTTP and HTTPS? Why is HTTPS important when you are using a banking or payment website?
  -  HTTP (HyperText Transfer Protocol):
      Data is sent in plain text
      Not secure
      Anyone can potentially read the data

      HTTPS (HyperText Transfer Protocol Secure):
      Data is encrypted (coded)
      Secure
      Uses SSL/TLS to protect information

      Why HTTPS is important for banking/payment websites

      When you use a banking or payment site, you enter:

      Passwords
      ATM/card details
      Personal information

      HTTPS protects this data by:

      Encrypting information so hackers can’t read it Preventing data theft and fraud
      Ensuring you are connected to the real website, not a fake one
